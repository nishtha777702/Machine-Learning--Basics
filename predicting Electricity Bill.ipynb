```python
## import necessay library 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Define the dataset
daily_usage_hours = np.array([[3], [4], [6], [7], [8], [10], [12]])  # Independent variable (X)
monthly_bills = np.array([50, 60, 80, 95, 110, 130, 150])            # Dependent variable (y)


# Step 2: Initialize and train the linear regression model
model = LinearRegression()
model.fit(daily_usage_hours, monthly_bills)

# Step 3: Predict the bill for a household that uses 5 hours per day
new_input = np.array([[5]])
predicted_bill = model.predict(new_input)

# Step 4: Get slope and intercept
slope = model.coef_[0]       # This tells us how much the bill increases per hour
intercept = model.intercept_ ##Intercept is the base cost (even when usage is 0).

# Step 5: Generate predictions for visualization
x_range = np.linspace(2, 13, 100).reshape(-1, 1)
y_pred_line = model.predict(x_range)

# Step 6: Plot the results
plt.scatter(daily_usage_hours, monthly_bills, color='blue', label='Actual Data')
plt.plot(x_range, y_pred_line, color='red', label='Regression Line')
plt.scatter(new_input, predicted_bill, color='green', label=f'Prediction (5 hrs): ${predicted_bill[0]:.2f}') #This adds a label for the dot, showing the predicted amount in dollars, nicely formatted to 2 decimal places
plt.title('Monthly Electricity Bill Prediction')
plt.xlabel('Average Daily Usage (hours)')
plt.ylabel('Monthly Bill ($)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

(predicted_bill[0], slope, intercept)
```

**What is the expected monthly bill for 5 hours/day usage?
 $72.01

**How much does the bill increase for each extra hour of usage?
About $11.40 per hour (this is the slope of the line)

**Intercept (Base fixed cost):
About $15.02 — this is the estimated base cost even if no electricity is used
