```python
## import the library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Load the dataset
df = pd.read_csv("Housing.csv")  # Make sure the file is in the same directory or provide the full path

# Step 2: Select features and target
X = df[['area']]  # Independent variable (house size)
y = df['price']   # Dependent variable (house price)

# Step 3: Initialize and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Step 4: Make predictions on training data
predictions = model.predict(X)

# Step 5: Predict price for a new house size (e.g., 1600 sq ft)
new_input = np.array([[3000]])
predicted_price = model.predict(new_input)
print(f"Predicted price for a house of size {new_input[0][0]} sq ft: ${predicted_price[0]:,.2f}")

# Step 6: Output model parameters
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)

# Step 7: Visualization
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction using Linear Regression')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
