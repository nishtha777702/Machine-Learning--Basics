```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # is the machine learning tool we’ll use

house_size = np.array([500, 650, 800, 900, 1000, 1100, 1200, 1350, 1500, 1600, 1700, 1800]).reshape(-1, 1)

# Corresponding prices ($1000s)
house_price = np.array([120, 150, 180, 200, 220, 240, 260, 285, 310, 330, 350, 370])
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Fit model
model = LinearRegression()   ### Hey machine, use LinearRegression() to learn the relationship between size and price.”
model.fit(house_size, house_price)  ###.fit() is where the model learns from the data.

# Predictions
predicted_price = model.predict(house_size) ###“Based on the house sizes I gave you, what prices do you think they should be?
```

```python
plt.scatter(house_size, house_price, color='blue', label='Actual Price')
plt.plot(house_size, predicted_price, color='red', label='Regression Line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($1000s)')
plt.title('House Price Prediction with Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
